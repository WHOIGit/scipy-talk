```python
%matplotlib inline
import numpy as np
from matplotlib import pyplot as plt
```

```python
def cdist(XA, XB):
    M, N = np.mgrid[:XA.shape[0],:XB.shape[0]]
    return np.sqrt(np.sum((XA[M] - XB[N])**2,axis=2))
```

```python
I, J = np.random.uniform(size=(2,10))
IJ = np.vstack((I,J)).T

plt.figure(figsize=(7,7))
plt.gca().set_aspect('equal')
plt.scatter(I,J)
```

```python
D = cdist(IJ,IJ)
D
```

```python
nn_ix = np.argsort(D,axis=1)
nn_ix
```

```python
k=3
knn_ix = nn_ix[:,:k]
knn_ix
```

```python
row = np.tile(np.arange(IJ.shape[0]).reshape(-1,1),k)
row
```

```python
knn_dist = D[row, knn_ix]
knn_dist
```

```python
def knn(query_pts, pts, k=3):
    # brute force k nearest neighbors
    D = cdist(query_pts, pts)
    knn_ix = np.argsort(D,axis=1)[:,:k]
    row = np.tile(np.arange(query_pts.shape[0]).reshape(-1,1),k)
    knn_dist = D[row, knn_ix]
    return knn_dist, knn_ix
```

```python
from scipy import stats

# now make a grid
resolution = 100
m = np.linspace(0, 1, resolution)
n = np.linspace(0, 1, resolution)
M, N = np.meshgrid(m,n)
MN = np.dstack((M,N)).reshape(-1,2)

# now find the k nearest neighbors for each grid point
dist, nabe = knn(MN,IJ,k=1)

dist_img = dist.reshape((resolution,resolution))

plt.figure(figsize=(7,7))
plt.gca().set_aspect('equal')
plt.scatter(I, J, c='black')
plt.imshow(0-dist_img,origin='lower',cmap='gray',extent=[0, 1, 0, 1])
```

```python
from skimage.color import label2rgb

label_img = nabe.reshape((resolution,resolution))
colors = np.random.uniform(low=0.25,size=(np.max(label_img)+1,3))

plt.figure(figsize=(7,7))
plt.gca().set_aspect('equal')
plt.scatter(I, J, c='black')
plt.imshow(label2rgb(label_img,colors=colors),interpolation='nearest',origin='lower',extent=[0, 1, 0, 1])
```
